```python
import pandas as pd

# Read Dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# First 5 Rows
print("FIRST 5 ROWS")
print(df.head())

# Last 5 Rows
print("\nLAST 5 ROWS")
print(df.tail())

# Data Types
print("\nDATA TYPES")
print(df.dtypes)

# Statistics
print("\nSUMMARY STATISTICS")
print(df.describe())

# Mean
print("\nMEAN")
print(df.mean(numeric_only=True))

# Median
print("\nMEDIAN")
print(df.median(numeric_only=True))

# Minimum Values
print("\nMINIMUM")
print(df.min(numeric_only=True))

# Maximum Values
print("\nMAXIMUM")
print(df.max(numeric_only=True))

# Filter Rows
filtered = df[df["sepal_length"] > 5]

print("\nFILTERED DATA")
print(filtered.head())

# Select Columns
selected = filtered[["sepal_length", "species"]]

# Save CSV
selected.to_csv("filtered_output.csv", index=False)

print("\nProject Completed Successfully!")
```
